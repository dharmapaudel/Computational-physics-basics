# Distance vs Acceleration  plot

```python
from math import *
def distance(u,t):
    a = 4
    s = u*t+ 0.5*a*t**2
    return [s]
```

```python
d = distance(100,t)
S =[]
a = []

for i in range(0,4):
    a.append(i)
for t in range(0,4):
    d = distance(100,t)
    S.append(d)
    

```

```python
import matplotlib.pyplot as plt
```
